## ETL

Extract-Transform-Load (ETL) - это причудливый способ сказать: "У нас есть некоторые грубые, унаследованные данные в этой системе, и теперь они нам нужны в этой блестящей новой системе, поэтому мы собираемся перенести их". (Как правило, за этим следует: «Нам нужно будет запустить это только один раз.» Затем за этим обычно следует много шлепков по лбу и стонов о том, насколько же это всё сложно.)

Цель 

Мы собираемся извлечь некоторые оценки из старой системы.   

Старая система хранила списки букв на соответствующие баллы:  
1 балл: «A», «E», «I», «O», «U», «L», «N», «R», «S», «T»,  
2 балла: «D», «G»,  
3 балла: «B», «C», «M», «P»,  
4 балла: «F», «H» ',' V ',' W ',' Y ',  
5 баллов:' K ',  
8 баллов:' J ',' X ',  
10 баллов:' Q ',' Z ',  

Новая блестящая система вместо этого хранит счет за букву, что позволяет намного быстрее и легче вычислить счет за слово. Он также сохраняет буквы в нижнем регистре независимо от регистра введенных букв:

Etc.

## Collatz Conjecture

Гипотеза Коллатца или проблема 3x + 1 может быть обобщена следующим образом: 

Возьмем любое натуральное число n. Если n четное, разделите n на 2, чтобы получить n / 2. Если n нечетно, умножьте n на 3 и добавьте 1, чтобы получить 3n + 1. 

Повторяйте процесс бесконечно. Предположение гласит, что независимо от того, с какого числа вы начнете, вы всегда в конечном итоге достигнете 1. Если задано число n, вернуть количество шагов, необходимое для достижения 1. 

*Пример* 

Начиная с n = 12, шаги будут следующими: `12 6 3 10 5 16 8 4 2 1` 

Итого 9 шагов. Таким образом, для входа n = 12 возвращаемое значение будет 9.

## Transpose

Грубо говоря, транспонирование матрицы: 
```
ABC 
DEF 
```
задается как:

Строки становятся столбцами, а столбцы становятся строками. См. Https://en.wikipedia.org/wiki/Transpose.

```
ABC 
DE
```

```
AD 
BE 
C 
```

```
AB 
DEF
```

## ISBN Verifier

Процесс проверки [ISBN-10](https://ru.wikipedia.org/wiki/Международный_стандартный_номер_книги) используется для проверки идентификационных номеров книг. 

Обычно они содержат тире и выглядят следующим образом: 3-598-21508-8 ISBN 

Формат ISBN-10 состоит из 9 цифр (от 0 до 9) плюс один контрольный символ (только цифра или только X). В случае, если контрольный символ представляет собой X, это представляет значение «10». Они могут быть переданы с дефисами или без них и могут быть проверены на их действительность по следующей формуле: 
```
(x1 * 10 x2 * 9 x3 * 8 x4 * 7 x5 * 6 x6 * 5 x7 * 4 x8 * 3 x9 * 2 x10 * 1) mod 11 == 0
```
Если результат равен 0, то это действительный ISBN-10, в противном случае он недействителен. 

Пример Давайте возьмем ISBN-10 3-598-21508-8. Мы подключаем его к формуле и получаем: 
```
(3 * 10 + 5 * 9 + 9 * 8 + 8 * 7 + 2 * 6 + 1 * 5 + 5 * 4 + 0 * 3 + 8 * 2 + 8 * 1) mod 11 == 0
```
Поскольку результат равен 0, это доказывает, что наш ISBN действителен. 

Задача Считывая строку, программа должна проверить, является ли предоставленная строка действительным ISBN-10. Чтобы реализовать это, нужно подумать о предварительной обработке / разборе строки перед вычислением контрольной цифры для ISBN. Программа должна быть в состоянии проверить ISBN-10 как с разделительными тире, так и без них. 

Предостережения Преобразование из строк с номерами может быть сложно на определенных языках. Теперь это еще сложнее, поскольку контрольная цифра ISBN-10 может быть «X» (обозначая «10»). Например, 3-598-21507-X является действительным ISBN-10. 

Бонусные задачи Реализация show (io :: IO, isbn :: ISBN) для печати ISBN в стандартном формате с тире. Создание действительного ISBN-13 из входного ISBN-10 (и, возможно, проверьте его снова с помощью производного верификатора). Создайте действительный ISBN, возможно, даже из заданного исходного ISBN.

## Complex Numbers

Комплексное число - это число в форме `a + b * i`, где a и b действительные и i удовлетворяет `i^2 = -1`. a называется действительной частью, а b называется мнимой частью z. 

Сопряженным числом `a + b * i` является число `a - b * i`. Абсолютное значение комплексного числа `z = a + b * i` является действительным числом `| z | = sqrt (a^2 + b^2)`. Квадрат абсолютного значения `|z|^2` является результатом умножения z на его комплексное сопряжение. 

Сумма / разница двух комплексных чисел включает в себя сложение / вычитание их действительной и мнимой частей 

Результат умножения по определению `(a + i * b ) * (c + i * d) = (a * c - b * d) (b * c + a * d) * i`. 

Взаимное значение ненулевого комплексного числа равно `1 / (a + i * b) = a/(a^2 + b^2) - b/(a^2 + b^2) * i.`. 

Деление комплексного числа `a + i*b` на другое `c + i*d` дает: ` (a + i * b) / (c + i * d) = (a * c + b * d)/(c^2 + d^2) + (b * c - a * d)/(c^2 + d^2) * i`. 

Возведение $e$ в комплексную степень можно выразить как `e^(a + i * b) = e^a * e^(i * b)`, последний член которого дается формулой Эйлера `e^(i * b) = cos (b) + i * sin (b)`. 

Реализуйте следующие операции: сложение, вычитание, умножение и деление двух комплексных чисел, сопряжение, абсолютное значение, экспонента заданного комплексного числа. 

Предположим, что используемый вами язык программирования не имеет реализации комплексных чисел. Реализацию комплексных чисел в Julia Base можно найти здесь: https://github.com/JuliaLang/julia/blob/master/base/complex.jl. Реализуйте `jm` аналогично `im` так, чтобы `1+1jm == ComplexNumber (1, 1)`.
