Генетическим языком всех живых существ на планете является ДНК. ДНК - это большая молекула, которая построена из чрезвычайно длинной последовательности отдельных элементов, называемых нуклеотидами. ## Hamming

рассчитать расстояние Хэмминга между двумя нитями ДНК. 

Твое тело состоит из клеток, которые содержат ДНК. Эти клетки регулярно изнашиваются и нуждаются в замене, чего они достигают, делясь на дочерние клетки. Фактически, средний человеческий организм испытывает около 10 квадриллионов клеточных делений в течение жизни! Когда клетки делятся, их ДНК тоже размножается. Иногда во время этого процесса происходят ошибки, и отдельные фрагменты ДНК кодируются с неверной информацией. 

Если мы сравним две нити ДНК и посчитаем разницу между ними, мы увидим, сколько ошибок произошло. Это известно как «Расстояние Хэмминга». Мы читаем ДНК, используя буквы C, A, G и T. Две нити могут выглядеть следующим образом: 
```
GAGCCTACTAACGGGAT 
CATCGTAATGACGGCCT 
^ ^ ^  ^ ^    ^^
```
Они имеют 7 отличий, поэтому расстояние Хэмминга равно 7. Расстояние Хэмминга полезно для многих вещей в науке, а не только для биологии, так что это хорошая фраза, с которой нужно ознакомиться :) 

Замечания по реализации Расстояние Хэмминга определено только для последовательностей равной длины, поэтому попытка рассчитать его между последовательностями разной длины не должно работать. Общая обработка этой ситуации (например, повышение исключения по сравнению с возвратом специального значения) может отличаться в разных языках.

## RNA Transcription

Получив цепочку ДНК, верните ее РНК-комплемент (согласно транскрипции РНК). 

Нити ДНК и РНК представляют собой последовательность нуклеотидов. Четырьмя нуклеотидами в ДНК являются аденин (A), цитозин (C), гуанин (G) и тимин (T). 

Четырьмя нуклеотидами в РНК являются аденин (A), цитозин (C), гуанин (G) и урацил (U). Учитывая цепочку ДНК, ее транскрибируемая цепочка РНК формируется путем замены каждого нуклеотида его комплементом:
```
G -> C
C -> G
T -> A
A -> U
```

## Run Length Encoding

Реализация кодирования и декодирования длин серий. 

Кодировка длин серий (RLE) - это простая форма сжатия данных, при которой прогоны (последовательные элементы данных) заменяются только одним значением данных и количеством. 

Например, мы можем ужать исходные 53 символа в 13. 
```
'WWWWWWWWWWWWWWWWWWWWWWWWWBWBWWWBBBWWWWWWWWWWWWWWWWWWWWWWWWWWWWWWWB' -> '12WB12W3B24WB'
```
Данные могут быть легко восстановлены. 
```
'AABCCCDEEEE' -> '2AB3CD4E' -> 'AABCCCDEEEE' 
```
Для простоты можно предположить, что незашифрованная строка будет содержать только буквы от A до Z (в нижнем или верхнем регистре) и пробел. Таким образом, данные, которые должны быть закодированы, никогда не будут содержать никаких цифр и цифр внутри данных, которые должны быть декодированы, всегда представляют счет для следующего символа.

## Binary Search

Реализовать алгоритм двоичного поиска. 

Поиск отсортированной коллекции - обычная задача. Словарь - это отсортированный список определений слов. По данному слову можно найти его определение. Телефонная книга - это отсортированный список имен, адресов и телефонных номеров людей. Зная чье-то имя, можно быстро найти его номер телефона и адрес. 

Если список для поиска содержит более нескольких элементов (скажем, дюжину), бинарный поиск потребует гораздо меньше сравнений, чем линейный поиск, но налагает требование сортировки списка. В информатике алгоритм бинарного поиска или поиска с половинным интервалом находит позицию указанного входного значения (поисковый «ключ») в массиве, отсортированном по значению ключа. На каждом шаге алгоритм сравнивает значение ключа поиска со значением ключа среднего элемента массива. Если ключи совпадают, то соответствующий элемент найден и возвращается диапазон индексов, равных значению ключа поиска. В противном случае, если ключ поиска меньше, чем ключ среднего элемента, алгоритм повторяет свое действие на подмассиве слева от среднего элемента или, если ключ поиска больше, на подмассиве до право. Если оставшийся массив для поиска пуст, ключ не может быть найден в массиве и возвращается специальное указание «не найден». 

В этом случае методы поиска в Julia обычно возвращают пустой диапазон, расположенный в точке вставки. Бинарный поиск делит число элементов на каждую итерацию пополам, поэтому поиск элемента (или определение его отсутствия) занимает логарифмическое время. Бинарный поиск - это дихотомический алгоритм поиска «разделяй и властвуй». Для упрощения можно предположить, что все элементы списка, которые нужно найти, уникальны. Не стесняйтесь реализовать решение, которое работает на списках с неуникальными элементами в качестве бонусной задачи.

## Rotational Cipher

Создайте реализацию ротационного шифра, также иногда называемого шифром Цезаря. 

Шифр Цезаря - это простой шифр сдвига, который полагается на транспонирование всех букв в алфавите с использованием целочисленной клавиши от 0 до 26. Использование клавиши 0 или 26 всегда даст одинаковый результат из-за модульной арифметики. Буква сдвигается на столько же значений, сколько значение ключа. Общими обозначениями для вращательных шифров является ROT + <ключ>.

Наиболее часто используемый ротационный шифр - ROT13. ROT13 для латинского алфавита будет выглядеть следующим образом: ```Обычный: abcdefghijklmnopqrstuvwxyz 
Шифр: nopqrstuvwxyzabcdefghijklm ```
Он сильнее, чем шифр Atbash, потому что он имеет 27 возможных ключей и 25 используемых ключей. Зашифрованный текст записывается в том же формате, что и ввод, включая пробелы и знаки препинания.
```
Examples
ROT5 omg gives trl
ROT0 c gives c
ROT26 Cool gives Cool
ROT13 The quick brown fox jumps over the lazy dog. gives Gur dhvpx oebja sbk whzcf bire gur ynml qbt.
ROT13 Gur dhvpx oebja sbk whzcf bire gur ynml qbt. gives The quick brown fox jumps over the lazy dog.
````

*Bonus A*  
Реализуйте строковый литерал, который действует как ROT13 в строке:

```R13"abcdefghijklmnopqrstuvwxyz" == "nopqrstuvwxyzabcdefghijklm"```

*Bonus B*  
Реализуйте строковые литералы R<i>, i = 0, ..., 26, которые сдвигают строку для значений i:
```
R0"Hello, World!" == "Hello, World!"
R4"Testing 1 2 3 testing" == "Xiwxmrk 1 2 3 xiwxmrk"
R13"abcdefghijklmnopqrstuvwxyz" == "nopqrstuvwxyzabcdefghijklm"
```

## Atbash Cipher

Создайте реализацию шифра atbash, древней системы шифрования, созданной на Ближнем Востоке. 

Шифр Atbash - это простой заменительный шифр, который полагается на транспонирование всех букв в алфавите так, чтобы полученный алфавит был обратным. Первая буква заменяется последней буквой, вторая - второй, последней и т.д. Шифр Atbash для латинского алфавита будет выглядеть следующим образом: 
```
Plain: abcdefghijklmnopqrstuvwxyz 
Cipher: zyxwvutsrqponmlkjihgfedcba 
```
Это очень слабый шифр, поскольку он имеет только один возможный ключ, и это простой одноалфавитный шифр замещения. Однако, это не могло быть проблемой во времена шифра. Зашифрованный текст записывается в группах фиксированной длины, традиционный размер группы составляет 5 букв, и знаки препинания исключаются. Это делается для того, чтобы было сложнее угадывать вещи на основе границ слов.

Examples
```
Encoding test gives gvhg
Decoding gvhg gives test
Decoding gsvjf rxpyi ldmul cqfnk hlevi gsvoz abwlt gives thequickbrownfoxjumpsoverthelazydog
```

## Difference Of Squares

Найти разность между квадратом суммы и суммой квадратов первых N натуральных чисел. 

Квадрат суммы первых десяти натуральных чисел равен (1 + 2 + ... + 10)² = 55² = 3025. 

Сумма квадратов первых десяти натуральных чисел составляет 1² + 2² +. .. + 10² = 385. 

Следовательно, разница между квадратом суммы первых десяти натуральных чисел и суммой квадратов первых десяти натуральных чисел составляет 3025 - 385 = 2640. 

От вас не ожидается найти эффективное решение этого самостоятельно из первых принципов; исследования поощряются. Поиск лучшего алгоритма для этой проблемы является ключевым навыком в разработке программного обеспечения.

## Pascal's Triangle

Вычислить треугольник Паскаля до заданного числа строк. 

В треугольнике Паскаля каждое число вычисляется путем сложения чисел справа и слева от текущей позиции в предыдущей строке.

```
    1
   1 1
  1 2 1
 1 3 3 1
1 4 6 4 1
# ... etc
```

## Raindrops

Преобразуйте число в строку, содержание которой зависит от делителей. 

* Если число имеет делитель 3, выведите «Pling». 
* Если число имеет делитель 5, выведите «Plang». 
* Если число имеет делитель 7, выведите «Plong». 
* Если число не имеет 3, 5 или 7 среди делителей, просто пропустите это число. 

Примеры 28: 1, 2, 4, 7, 14, 28. Если говорить по каплям, это будет простой «Plong». 

 30: 1, 2, 3, 5, 6, 10, 15, 30. В языке дождевой капли это будет «PlingPlang». 

У 34 есть четыре делителя: 1, 2, 17 и 34. В языке дождевой капли это будет «34».

## Scrabble Score

По заданному слову вычислите оценку этого слова

```
Letter                           Value
A, E, I, O, U, L, N, R, S, T       1
D, G                               2
B, C, M, P                         3
F, H, V, W, Y                      4
K                                  5
J, X                               8
Q, Z                               10
```

Примеры  
"cabbage" наберет 14 очков:
```
3 points for C
1 point for A, twice
3 points for B, twice
2 points for G
1 point for E
And to total:

3 + 2*1 + 2*3 + 2 + 1
= 3 + 2 + 6 + 3
= 5 + 9
= 14
```

## Word Count

Для данной фразы подсчитайте вхождения каждого слова в этой фразе.

Например, для ввода «olly olly in come free»   
olly: 2 in: 1 come: 1 free: 1 

## Luhn

По заданному числу определите, является ли оно действительным по [формуле Луна](https://ru.wikipedia.org/wiki/Алгоритм_Луна). 

Алгоритм Луна - это простая формула контрольной суммы, используемая для проверки различных идентификационных номеров, таких как номера кредитных карт и канадские номера социального страхования. Суть задачи состоит в проверке, является ли данная строка действительной. Проверка числа Строки длиной 1 или меньше недопустимы. Пробелы разрешены на входе, но они должны быть удалены перед проверкой. Все другие нецифровые символы запрещены. 

*Пример 1:*  

действительный номер кредитной карты `4539 1488 0343 6467 `

Первым шагом алгоритма Луна является удвоение каждой второй цифры, начиная с правой. 

Мы будем удваивать `4_3_ 1_8_ 0_4_ 6_6_` 

Если при удвоении числа получится число больше 9, то вычтите 9 из произведения. 

Результаты нашего удвоения: `8569 2478 0383 3437 `

Затем сложите все цифры: `8 5 6 9 2 4 7 8 0 3 8 3 3 4 3 7 = 80`  

Если сумма делится равномерно на 10, тогда номер действителен. Этот номер действителен! 

*Пример 2:* неверный номер кредитной карты `8273 1232 7352 0569`  

Удвоить вторые цифры, начиная с правого `7253 2262 5312 0539` 

Суммировать цифры `7 2 5 3 2 2 6 2 5 3 1 2 0 5 3 9 = 57` 

57 не делится поровну на 10, поэтому это число недопустимо.

## Roman Numerals

Напишите функцию для преобразования нормальных чисел в римские. 

Римляне были умными. Они завоевали большую часть Европы и правили ею сотни лет.Они изобрели бетонные и прямые дороги и даже бикини. Единственное, что они никак не обнаружили, так это нулевое число. Это сделало написание и датирование обширных историй их подвигов немного более сложным, но система чисел, которую они придумали, все еще используется сегодня. 

Например, BBC использует римские цифры для датировки своих программ. Римляне написали цифры, используя буквы - I, V, X, L, C, D, M. (обратите внимание, что эти буквы имеют много прямых линий и, следовательно, их легко выбить на каменных табличках). 1 => I 10 => X 7 => VII Нет необходимости в преобразовании чисел, превышающих 3000. (Сами римляне не склонны идти выше) 

Википедия говорит: Современные римские цифры ... пишутся, выражая каждую цифру отдельно, начиная с самой левой цифры и пропуская любую цифру со значением ноль. 

В римских цифрах 1990 означает MCMXC: 1000 = M 900 = CM 90 = XC 

2008 записывается как MMVIII: 2000 = MM 8 = VIII 

Смотрите также: http://www.novaroma.org/via_romana/numbers.html

## Isogram

Определите, является ли слово или фраза изограммой. 

Изограмма (также известная как «непатентованное слово») - это слово или фраза без повторяющейся буквы, однако пробелы и дефисы могут появляться несколько раз. 

Примеры изограмм: 
* lumberjacks 
* background 
* downstream
Слово изограммы, однако, не является изограммой, потому что 'м' повторяется.

Source
Wikipedia https://en.wikipedia.org/wiki/Isogram
